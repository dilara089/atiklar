```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2

def change_back(background, img):


    x, y = 0, 0
    background = cv2.resize(background, (img.shape[1], img.shape[0]), interpolation = cv2.INTER_AREA)
    res = np.copy(background)
    place = res[y: y + img.shape[0], x: x + img.shape[1]]
    a = img[..., 3:].repeat(3, axis=2).astype('uint16')
    place[...] = (place.astype('uint16') * (255 - a) // 255) + img[..., :3].astype('uint16') * a // 255
    return res
img_path= "/content/output.png"
back_path="/content/deniz.jpg"

img = cv2.imread(img_path,cv2.IMREAD_UNCHANGED)
back = cv2.imread(back_path)

result = change_back(back,img)
cv2.imwrite("result.png",result)
```

```python
import cv2
image = cv2.imread("/content/result.png")
# Metin dizesi
text = "vacation baby"
# Yaz tipi
font = cv2.FONT_HERSHEY_SIMPLEX
# Yaz tipi boyutu
font_scale = 1
# RGB formatnda renk (Red, Green, Blue).
# Beyaz renk
color = (255, 255, 255)
# izginin kalnl
thickness = 2
x, y = 100, 100
image_with_text = cv2.putText(image, text, (x, y), font, font_scale, color, thickness, cv2.LINE_AA)
```
